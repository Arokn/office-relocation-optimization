# PARTIE 1 : Linear programming relaxation

### Définition des Ensembles

* **Groupes (Services) :** $G = \{P, M, T, S, O\}$
* **Bureaux :** $Bu = \{A1, A2, B1, B2, B3, C1, C2, D1, D2, D3, E1, E2\}$
* **Ailes du bâtiment :**
  * $\text{Aile A} = \{A1, A2\}$
  * $\text{Aile B} = \{B1, B2, B3\}$
  * $\text{Aile C} = \{C1, C2\}$
  * $\text{Aile D} = \{D1, D2, D3\}$
  * $\text{Aile E} = \{E1, E2\}$

### Définition des variables de décision

* **$x_{i,j,p}^g \ge 0$** : Le nombre de personnes du groupe $g$ qui se déplacent du bureau $i$ vers le bureau $j$ en phase $p$.
* **$\text{occ}_{i,p}^g \ge 0$** : Le nombre de personnes du groupe $g$ qui occupent le bureau $i$ en phase $p$.

*Avec $g \in G$ (l'ensemble des services/groupes).*

### Fonction objectif

L'objectif est de minimiser l'effort global de transfert, ce qui correspond au nombre total de déménagements de personnel tout au long du processus :

$$\min \sum_{p \in P} \sum_{\substack{g \in G \\ i \in Bu \\ j \in Bu \\ i \neq j}} x_{i,j,p}^g$$

### Contraintes du modèle

* **Conservation des flux (Dynamique de l'occupation) :**
  Pour chaque bureau $i$, chaque groupe $g$, et pour les phases $1 \le p \le 5$ :
  $$\text{occ}_{i,p}^g = \text{occ}_{i,p-1}^g - \sum_{\substack{j \in Bu \\ j \neq i}} x_{i,j,p}^g + \sum_{\substack{j \in Bu \\ j \neq i}} x_{j,i,p}^g$$

* **Capacité maximale des bureaux :**
  Le nombre total de personnes (tous groupes confondus) dans un bureau $j$ à une phase $p$ donnée ne peut pas dépasser 2:
  $$\sum_{g \in G} \text{occ}_{j,p}^g \le 2 \quad \forall j \in Bu, \forall p \in P$$

* **Cohérence des départs :**
  On ne peut pas faire partir d'un bureau plus de personnes d'un groupe $g$ qu'il n'y en avait à la phase précédente :
  $$\sum_{\substack{j \in Bu \\ j \neq i}} x_{i,j,p}^g \le \text{occ}_{i,p-1}^g \quad \forall i \in Bu, \forall p \in P, \forall g \in G$$

*Avec $i, j \in Bu$ (l'ensemble des bureaux).*

* **Contraintes de rénovation (Indisponibilité des locaux) :**
  Lorsqu'une aile est en cours de rénovation pendant une phase donnée, aucun bureau de cette aile ne peut être occupé. De plus, aucun flux de déménagement ne peut avoir pour destination un bureau en travaux.

  * **Phase 1 (Rénovation de l'aile B) :** 
    $$\text{occ}_{i,1}^g = 0 \quad \forall i \in \text{AileB}, \forall g \in G$$
  * **Phase 2 (Rénovation de l'aile D) :** 
    $$\text{occ}_{i,2}^g = 0 \quad \forall i \in \text{AileD}, \forall g \in G$$
  * **Phase 3 (Rénovation de l'aile C) :** 
    $$\text{occ}_{i,3}^g = 0 \quad \forall i \in \text{AileC}, \forall g \in G$$
  * **Phase 4 (Rénovation de l'aile A) :** 
    $$\text{occ}_{i,4}^g = 0 \quad \forall i \in \text{AileA}, \forall g \in G$$

  * **Interdiction d'emménagement en zone de travaux :**
    Si un bureau $j$ est situé dans une aile en rénovation à la phase $p$, alors aucun déplacement vers ce bureau n'est autorisé:
    $$\sum_{i \in Bu} \sum_{g \in G} x_{i,j,p}^g = 0$$

### Multiplicateurs de Lagrange (Variables Duales)

Pour construire le problème dual et formuler les conditions KKT, nous associons à chaque contrainte du problème primal un multiplicateur de Lagrange :

* **$\lambda_{i,p}^g \in \mathbb{R}$** : Associé aux contraintes d'égalité de la conservation des flux (équilibre des flux pour le groupe $g$ au bureau $i$ à la phase $p$).
* **$\mu_{i,p} \ge 0$** : Associé aux contraintes d'inégalité pour la capacité maximale de 2 personnes par bureau.
* **$\gamma_{i,p}^g \ge 0$** : Associé aux contraintes de cohérence des départs (limitation des flux sortants par l'occupation réelle).
* **$\alpha_j^{g_0}, \beta_j^{g_F}$** : Associés aux contraintes d'état initial (phase 0) et d'état final (phase 5).

### Conditions d'Optimalité KKT & Dualité

* **Pour la variable $x_{i,j,p}^g$ :**
  $$1 + \lambda_{j,p}^g - \lambda_{i,p}^g + \gamma_{i,p}^g - a_{i,j,p}^g = 0$$

* **Pour la variable $\text{occ}_{j,p}^g$ :**
  $$\lambda_{j,p}^g - \lambda_{j,p+1}^g + \mu_{j,p} - \gamma_{j,p+1}^g - b_{j,p}^g = 0$$

---

### Primal :
* $x_{i,j,p}^g \ge 0 \quad , \quad \text{occ}_{j,p}^g \ge 0$
* $\sum_{g \in G} \text{occ}_{j,p}^g \le 2$
* $\sum_{j \neq i} x_{i,j,p}^g \le \text{occ}_{i,p-1}^g$
* $\text{occ}_{j,p}^g - \text{occ}_{j,p-1}^g - \sum_{i \neq j} x_{i,j,p}^g + \sum_{k \neq j} x_{j,k,p}^g = 0$

---

### Dual :
$$\mu_{j,p} \ge 0 \quad , \quad \gamma_{i,p}^g \ge 0 \quad , \quad a_{i,j,p}^g \ge 0 \quad , \quad b_{j,p}^g \ge 0$$

---

### Conditions de complémentarité (Écarts nuls) :
* $-\ \mu_{j,p} \times \left(2 - \sum_{g \in G} \text{occ}_{j,p}^g\right) = 0$
* $-\ \gamma_{i,p}^g \times \left(\text{occ}_{i,p-1}^g - \sum_{j} x_{i,j,p}^g\right) = 0$
* $-\ a_{i,j,p}^g \times x_{i,j,p}^g = 0$
* $-\ b_{j,p}^g \times \text{occ}_{j,p}^g = 0$

### Implémentation Numérique avec PuLP

Pour résoudre ce programme linéaire (LP), nous traduisons la formulation mathématique en code Python en utilisant la bibliothèque de recherche opérationnelle **PuLP**. Ce solveur va nous permettre de modéliser explicitement nos variables de décision, notre fonction objectif ainsi que l'ensemble des contraintes de flux et de capacité afin de trouver la configuration de déménagement optimale.

```python
import pulp
```

```python
offices = [
    'A1', 'A2', 
    'B1', 'B2', 'B3', 
    'C1', 'C2', 
    'D1', 'D2', 'D3', 
    'E1', 'E2'
]

wings = {
    'A': ['A1', 'A2'],
    'B': ['B1', 'B2', 'B3'],
    'C': ['C1', 'C2'],
    'D': ['D1', 'D2', 'D3'],
    'E': ['E1', 'E2']
}

groups = ['M', 'P', 'O', 'T', 'S']
phases = [1, 2, 3, 4, 5] 

renovation_schedule = {
    1: 'B',
    2: 'D',
    3: 'C',
    4: 'A',
    5: None 
}

# POur les données initiales de l'enoncé 
initial_occ = {o: {g: 0 for g in groups} for o in offices}
initial_occ['A1']['P'] = 2
initial_occ['A2']['P'] = 2
initial_occ['B1']['O'] = 1; initial_occ['B1']['M'] = 1
initial_occ['B2']['S'] = 2
initial_occ['B3']['O'] = 1; initial_occ['B3']['T'] = 1
initial_occ['C1']['T'] = 2
initial_occ['C2']['S'] = 2
initial_occ['D1']['T'] = 1
initial_occ['D2']['M'] = 1
initial_occ['D3']['M'] = 2


# Pour la phase final 

target_occ = {o: {g: 0 for g in groups} for o in offices}
target_occ['A1']['P'] = 2; target_occ['A2']['P'] = 2
target_occ['B1']['M'] = 1; target_occ['B2']['M'] = 2; target_occ['B3']['M'] = 1
target_occ['C1']['S'] = 2; target_occ['C2']['S'] = 2
target_occ['D1']['T'] = 1; target_occ['D2']['T'] = 2; target_occ['D3']['T'] = 1
target_occ['E1']['O'] = 1; target_occ['E2']['O'] = 1
```

```python
prob = pulp.LpProblem("Office_Relocation_Min_Moves", pulp.LpMinimize)


# On définit les variabes 
#ici la variable x^g: i,j,p 
move_vars = {}
for p in phases:
    for g in groups:
        for i in offices:
            for j in offices:
                if i != j:
                    move_vars[(p, g, i, j)] = pulp.LpVariable(f"x_{p}_{g}_{i}_{j}", 0, None, pulp.LpContinuous)


#ici on définit la variable occ^g :i,p
occ_vars = {}
all_phases_inc_0 = [0] + phases 

for p in all_phases_inc_0:
    for g in groups:
        for j in offices:
            occ_vars[(p, g, j)] = pulp.LpVariable(f"occ_{p}_{g}_{j}", 0, 2, pulp.LpContinuous)



#FONCTION BUT 
prob += pulp.lpSum([move_vars[(p, g, i, j)] for p, g, i, j in move_vars]) #but minimiser les deplacement 



# Etat initiale 
for g in groups:
    for j in offices:
        prob += occ_vars[(0, g, j)] == initial_occ[j][g], f"Init_Phase0_{j}_{g}"

# 
for p in phases:
    # contrainte 1 Occupants en P = Occupants en P-1 - Départs + Arrivées
    for g in groups:
        for j in offices:
            inflow = pulp.lpSum([move_vars[(p, g, i, j)] for i in offices if i != j])
            outflow = pulp.lpSum([move_vars[(p, g, j, k)] for k in offices if k != j])
            prob += occ_vars[(p, g, j)] == occ_vars[(p-1, g, j)] - outflow + inflow, f"Flow_{p}_{j}_{g}"

    # contraintes 2 :  deux personnes par bureau max 
    for j in offices:
        prob += pulp.lpSum([occ_vars[(p, g, j)] for g in groups]) <= 2, f"Cap_{p}_{j}"

    # contrainte 3 :  On ne peut pas déplacer plus de gens qu'il n'y en a 
    for g in groups:
        for i in offices:
            total_leaving_i = pulp.lpSum([move_vars[(p, g, i, j)] for j in offices if j != i])
            prob += total_leaving_i <= occ_vars[(p-1, g, i)], f"LimitMove_{p}_{i}_{g}"

    # Les Rénovations 
    wing_closed = renovation_schedule[p]
    if wing_closed:
        closed_offices = wings[wing_closed]
        for j in closed_offices:
            # Le bureau en renovation doit être vide à la fin de la phase
            for g in groups:
                prob += occ_vars[(p, g, j)] == 0, f"Reno_Closed_{p}_{j}_{g}"
            
            #on interdit les mouvements VERS un bureau en rénovation
            for g in groups:
                for i in offices:
                    if i != j:
                         prob += move_vars[(p, g, i, j)] == 0, f"No_Entry_Reno_{p}_{i}_{j}_{g}"
                         
# Etat final 
for g in groups:
    for j in offices:
        prob += occ_vars[(5, g, j)] == target_occ[j][g], f"Target_Phase5_{j}_{g}"
        
```

```python
status = prob.solve()

print("Status:", pulp.LpStatus[status])
print("Total deplacements:", pulp.value(prob.objective))

if pulp.LpStatus[status] == "Optimal":
    for p in phases:
        print("Phase", p)
        for g in groups:
            for i in offices:
                for j in offices:
                    if i != j:
                        val = pulp.value(move_vars[(p, g, i, j)])
                        if val is not None and val > 0.001:
                            print(p, g, i, j, int(round(val)))
```

### Résultats et Observations

Le programme linéaire (LP) est réalisable et converge vers une solution optimale.

* **Faisabilité :** Le LP semble réalisable.
* **OBSERVATION :** Le nombre minimal de déplacements est 30.

```python
if pulp.LpStatus[status] == "Optimal":
    dual_solutions = {
        "Lambda": [],
        "Mu": [],
        "Gamma": [],
        "Autres": []
    }

    for name, constraint in prob.constraints.items():
        pi = constraint.pi
        if pi is not None and abs(pi) > 1e-5:
            if "Flow" in name:
                dual_solutions["Lambda"].append((name, pi))
            elif "Cap" in name:
                dual_solutions["Mu"].append((name, pi))
            elif "LimitMove" in name:
                dual_solutions["Gamma"].append((name, pi))
            else:
                dual_solutions["Autres"].append((name, pi))

    for category, items in dual_solutions.items():
        for name, val in items:
            print(name, round(val, 4))
```

## Contraintes de Cohabitation Institutionnelle

Pour affiner le modèle, nous intégrons des contraintes de gestion spécifiques concernant la Présidence ($P$) et l'Association Étudiante ($S$).

### 1. Homogénéité des Bureaux
Les membres de la Présidence et de l'Association Étudiante ne cohabitent pas avec d'autres services. Un bureau contenant un membre de l'un de ces groupes leur est exclusivement réservé.

* **Modélisation :** Pour tout bureau $j$ et phase $p$ :
  $$\text{si } \text{occ}_{j,p}^P > 0 \implies \text{occ}_{j,p}^g = 0 \quad \forall g \neq P$$
  $$\text{si } \text{occ}_{j,p}^S > 0 \implies \text{occ}_{j,p}^g = 0 \quad \forall g \neq S$$

### 2. Isolation Géographique (Séparation des Ailes)
Pour éviter la proximité directe, les bureaux de l'Association Étudiante ne doivent pas être situés sur la même aile que ceux de la Présidence.

* **Modélisation :** Pour chaque aile et chaque phase $p$ :
  $$\left(\sum_{i \in \text{Aile}} \text{occ}_{i,p}^P\right) \times \left(\sum_{j \in \text{Aile}} \text{occ}_{j,p}^S\right) = 0 \quad \forall \text{Aile} \in \{A, B, C, D, E\}$$

### Impact sur le Modèle
* **Faisabilité :** Le programme linéaire reste réalisable.
* **OBSERVATION :** Le nombre minimal de déplacements reste fixé à **30**. L'arborescence des flux offre assez de flexibilité pour respecter l'isolation des ailes sans surcoût.

```python
#ON Crée le LP 
prob = pulp.LpProblem("Office_Relocation_Min_Moves", pulp.LpMinimize)

# On définit les variabes 


# Indicateurs de bureaux
z_P = pulp.LpVariable.dicts("z_P", ((p, j) for p in phases for j in offices), cat=pulp.LpBinary)
z_S = pulp.LpVariable.dicts("z_S", ((p, j) for p in phases for j in offices), cat=pulp.LpBinary)

# Indicateurs d'ailes
W_P = pulp.LpVariable.dicts("W_P", ((p, w) for p in phases for w in wings.keys()), cat=pulp.LpBinary)
W_S = pulp.LpVariable.dicts("W_S", ((p, w) for p in phases for w in wings.keys()), cat=pulp.LpBinary)


#ici la variable x^g: i,j,p 
move_vars = {}
for p in phases:
    for g in groups:
        for i in offices:
            for j in offices:
                if i != j:
                    move_vars[(p, g, i, j)] = pulp.LpVariable(f"x_{p}_{g}_{i}_{j}", 0, None, pulp.LpContinuous)


#ici on définit la variable occ^g :i,p
occ_vars = {}
all_phases_inc_0 = [0] + phases 

for p in all_phases_inc_0:
    for g in groups:
        for j in offices:
            occ_vars[(p, g, j)] = pulp.LpVariable(f"occ_{p}_{g}_{j}", 0, 2, pulp.LpContinuous)



#FONCTION BUT 
prob += pulp.lpSum([move_vars[(p, g, i, j)] for p, g, i, j in move_vars]) #but minimiser les deplacement 



# Etat initiale 
for g in groups:
    for j in offices:
        prob += occ_vars[(0, g, j)] == initial_occ[j][g], f"Init_Phase0_{j}_{g}"



for p in phases:
    # contrainte 1 Occupants en P = Occupants en P-1 - Départs + Arrivées
    for g in groups:
        for j in offices:
            inflow = pulp.lpSum([move_vars[(p, g, i, j)] for i in offices if i != j])
            outflow = pulp.lpSum([move_vars[(p, g, j, k)] for k in offices if k != j])
            prob += occ_vars[(p, g, j)] == occ_vars[(p-1, g, j)] - outflow + inflow, f"Flow_{p}_{j}_{g}"

    # contraintes 2 :  deux personnes par bureau max 
    for j in offices:
        prob += pulp.lpSum([occ_vars[(p, g, j)] for g in groups]) <= 2, f"Cap_{p}_{j}"

    # contrainte 3 :  On ne peut pas déplacer plus de gens qu'il n'y en a 
    for g in groups:
        for i in offices:
            total_leaving_i = pulp.lpSum([move_vars[(p, g, i, j)] for j in offices if j != i])
            prob += total_leaving_i <= occ_vars[(p-1, g, i)], f"LimitMove_{p}_{i}_{g}"

    # Les Rénovations 
    wing_closed = renovation_schedule[p]
    if wing_closed:
        closed_offices = wings[wing_closed]
        for j in closed_offices:
            # Le bureau en renovation doit être vide à la fin de la phase
            for g in groups:
                prob += occ_vars[(p, g, j)] == 0, f"Reno_Closed_{p}_{j}_{g}"
            
            #on interdit les mouvements VERS un bureau en rénovation
            for g in groups:
                for i in offices:
                    if i != j:
                         prob += move_vars[(p, g, i, j)] == 0, f"No_Entry_Reno_{p}_{i}_{j}_{g}"

    for j in offices:
      #pour P
       prob += occ_vars[(p, 'P', j)] <= 2 * z_P[(p, j)]                     
       other_than_P = [occ_vars[(p, g, j)] for g in groups if g != 'P']
       prob += pulp.lpSum(other_than_P) <= 2 * (1 - z_P[(p, j)])           
    
      #pour S
       prob += occ_vars[(p, 'S', j)] <= 2 * z_S[(p, j)]                     
       other_than_S = [occ_vars[(p, g, j)] for g in groups if g != 'S']
       prob += pulp.lpSum(other_than_S) <= 2 * (1 - z_S[(p, j)])           


    # Séparation par Aile
    for w_name, w_offices in wings.items():
        for j in w_offices:
            # Si le bureau j est P, l'aile est P
            prob += z_P[(p, j)] <= W_P[(p, w_name)], f"Wing_P_link_{p}_{w_name}_{j}"
            # Si le bureau j est S, l'aile est S
            prob += z_S[(p, j)] <= W_S[(p, w_name)], f"Wing_S_link_{p}_{w_name}_{j}"
        
        # L'aile ne peut pas avoir les deux
        prob += W_P[(p, w_name)] + W_S[(p, w_name)] <= 1, f"No_Mix_Wing_{p}_{w_name}"
                         
# Etat final 
for g in groups:
    for j in offices:
        prob += occ_vars[(5, g, j)] == target_occ[j][g], f"Target_Phase5_{j}_{g}"
        
```

### Enrichissement du Modèle : Variables Indicatrices

Pour intégrer proprement les règles d'exclusivité et d'isolation géographique sans introduire de non-linéarité, nous introduisons de nouvelles variables :

* **$Z_{j,p}^P$** : Variable indiquant si le bureau $j$ est attribué à la Présidence ($P$) à la phase $p$.
* **$Z_{j,p}^S$** : Variable indiquant si le bureau $j$ est attribué à l'Association Étudiante ($S$) à la phase $p$.
* **$W_{\text{Aile},p}^P$** : Variable indiquant si l'aile spécifiée abrite la Présidence ($P$) à la phase $p$.
* **$W_{\text{Aile},p}^S$** : Variable indiquant si l'aile spécifiée abrite l'Association Étudiante ($S$) à la phase $p$.

---

### Formulation des Nouvelles Contraintes

#### 1. Exclusivité et Homogénéité des Bureaux
Un membre de la Présidence (ou des Étudiants) ne peut partager son bureau qu'avec un membre de son propre service :
* $\text{occ}_{j,p}^g \le 2 \times (1 - Z_{j,p}^P) \quad \forall g \neq P$
* $\text{occ}_{j,p}^g \le 2 \times (1 - Z_{j,p}^S) \quad \forall g \neq S$

#### 2. Isolation des Ailes (Non-cohabitation)
La Présidence et l'Association Étudiante ne peuvent en aucun cas occuper la même aile au cours d'une même phase $p$ :
* $W_{\text{Aile},p}^P + W_{\text{Aile},p}^S \le 1 \quad \forall \text{Aile} \in \{A, B, C, D, E\}$

*Note : Les variables $W$ et $Z$ sont liées de sorte que si un bureau $j \in \text{Aile}$ a son indicateur $Z_{j,p} = 1$, alors l'indicateur d'aile correspondant $W_{\text{Aile},p}$ est activé.*

```python
status = prob.solve()
print("Status:", pulp.LpStatus[status])
print("Total deplacements:", pulp.value(prob.objective))

if pulp.LpStatus[status] == "Optimal":
    for p in phases:
        for g in groups:
            for i in offices:
                for j in offices:
                    if i != j:
                        val = pulp.value(move_vars[(p, g, i, j)])
                        if val is not None and val > 0.001:
                            print(p, g, i, j, int(round(val)))
```

### Analyse des Résultats du Modèle Restreint

L'intégration des contraintes d'exclusivité et d'isolement géographique préserve la faisabilité du problème.

* **Faisabilité :** Le programme linéaire reste réalisable.
* **OBSERVATION :** Le nombre minimal de déplacements passe à **31**. 

L'ajout de ces restrictions opérationnelles strictes (non-cohabitation et séparation des ailes) n'engendre qu'un surcoût mineur d'un unique déplacement supplémentaire par rapport au modèle initial.

# PARTIE 2 : Penalized problem

### Extension du Modèle : Linéarisation de la Pénalité Temporelle

Pour inciter le modèle à converger le plus tôt possible vers la configuration cible (finale), nous introduisons une pénalité basée sur l'écart entre l'occupation à une phase donnée et l'état final visé. 

Afin de linéariser la valeur absolue (norme $\ell_1$) de cet écart dans notre fonction objectif, nous introduisons une variable d'écart positive :

* **$\text{diff}_{j,p}^g \ge 0$** : Variable d'écart représentant la distance absolue entre l'occupation réelle du bureau $j$ par le groupe $g$ à la phase $p$ et son occupation cible finale.

---

### Formulation des Contraintes de Linéarisation

Pour chaque bureau $j$, chaque groupe $g$ et chaque phase $p$, la variable $\text{diff}_{j,p}^g$ est encadrée par les contraintes linéaires suivantes :

* $\text{diff}_{j,p}^g \ge \text{occ}_{j,p}^g - \text{Target}_{j}^g$
* $\text{diff}_{j,p}^g \ge \text{Target}_{j}^g - \text{occ}_{j,p}^g$

*Avec $\text{Target}_{j}^g$ représentant la configuration finale attendue pour le bureau $j$ (Phase 5).*

---

### Nouvelle Fonction Objectif Pénalisée

La fonction objectif intègre désormais le coût des déménagements cumulé à la pénalité globale pondérée par le paramètre $\lambda$ :

$$\min \sum_{p \in P} \sum_{\substack{g \in G \\ i \in Bu \\ j \in Bu \\ i \neq j}} x_{i,j,p}^g + \lambda \sum_{p \in P} \sum_{j \in Bu} \sum_{g \in G} \text{diff}_{j,p}^g$$

```python

prob = pulp.LpProblem("Office_Relocation_Dual_Q5", pulp.LpMinimize)
lambda_val = 100  


# Mouvements
move_vars = pulp.LpVariable.dicts("x", (phases, groups, offices, offices), 0, None, pulp.LpContinuous)
# Occupation
occ_vars = pulp.LpVariable.dicts("occ", (all_phases_inc_0, groups, offices), 0, 2, pulp.LpContinuous)
# Écart L1
diff_vars = pulp.LpVariable.dicts("diff", (all_phases_inc_0, groups, offices), 0, None, pulp.LpContinuous)
# Indicateurs Ailes/Bureaux 
z_P = pulp.LpVariable.dicts("z_P", (phases, offices), 0, 1, pulp.LpContinuous)
z_S = pulp.LpVariable.dicts("z_S", (phases, offices), 0, 1, pulp.LpContinuous)
W_P = pulp.LpVariable.dicts("W_P", (phases, wings.keys()), 0, 1, pulp.LpContinuous)
W_S = pulp.LpVariable.dicts("W_S", (phases, wings.keys()), 0, 1, pulp.LpContinuous)

# fct objectif 
total_moves = pulp.lpSum([move_vars[p][g][i][j] for p in phases for g in groups for i in offices for j in offices if i != j])
total_penalty = pulp.lpSum([diff_vars[p][g][j] for p in all_phases_inc_0 for g in groups for j in offices])
prob += total_moves + lambda_val * total_penalty


# A. État Initial et final 
for g in groups:
    for j in offices:
        prob += occ_vars[0][g][j] == initial_occ[j][g], f"Init_{g}_{j}"
        prob += occ_vars[5][g][j] == target_occ[j][g], f"Target_{g}_{j}"

# B. Linéarisation L1 
for p in all_phases_inc_0:
    for g in groups:
        for j in offices:
            prob += diff_vars[p][g][j] >= occ_vars[p][g][j] - target_occ[j][g], f"Abs_Pos_{p}_{g}_{j}"
            prob += diff_vars[p][g][j] >= target_occ[j][g] - occ_vars[p][g][j], f"Abs_Neg_{p}_{g}_{j}"

# C. Logique de Phase
for p in phases:
    # Conservation flux
    for g in groups:
        for j in offices:
            inflow = pulp.lpSum([move_vars[p][g][i][j] for i in offices if i != j])
            outflow = pulp.lpSum([move_vars[p][g][j][k] for k in offices if k != j])
            prob += occ_vars[p][g][j] == occ_vars[p-1][g][j] - outflow + inflow, f"Flow_{p}_{g}_{j}"

    # Capacité
    for j in offices:
        prob += pulp.lpSum([occ_vars[p][g][j] for g in groups]) <= 2, f"Cap_{p}_{j}"
    
    # Limite mouvement
    for g in groups:
        for i in offices:
            prob += pulp.lpSum([move_vars[p][g][i][j] for j in offices if j != i]) <= occ_vars[p-1][g][i], f"Lim_{p}_{g}_{i}"

    # Rénovations
    closed_wing = renovation_schedule[p]
    if closed_wing:
        for j in wings[closed_wing]:
            for g in groups:
                prob += occ_vars[p][g][j] == 0, f"Reno_{p}_{g}_{j}"
                for i in offices:
                    if i != j: prob += move_vars[p][g][i][j] == 0, f"NoIn_{p}_{g}_{i}_{j}"

    # Séparation P et S
    for j in offices:
        prob += occ_vars[p]['P'][j] <= 2 * z_P[p][j], f"Zp_ub_{p}_{j}"
        prob += pulp.lpSum([occ_vars[p][g][j] for g in groups if g != 'P']) <= 2 * (1 - z_P[p][j]), f"Zp_mix_{p}_{j}"
        prob += occ_vars[p]['S'][j] <= 2 * z_S[p][j], f"Zs_ub_{p}_{j}"
        prob += pulp.lpSum([occ_vars[p][g][j] for g in groups if g != 'S']) <= 2 * (1 - z_S[p][j]), f"Zs_mix_{p}_{j}"

    for w_name, w_offices in wings.items():
        for j in w_offices:
            prob += z_P[p][j] <= W_P[p][w_name], f"WingP_{p}_{w_name}_{j}"
            prob += z_S[p][j] <= W_S[p][w_name], f"WingS_{p}_{w_name}_{j}"
        prob += W_P[p][w_name] + W_S[p][w_name] <= 1, f"NoMix_{p}_{w_name}"


status = prob.solve()
print(f"Status: {pulp.LpStatus[status]}")
```

### Analyse de la Pénalisation dans le Problème Dual

L'ajout de la pénalité $L_1$ revient à minimiser :
$$\min z = \sum \text{Mouvements} + \lambda \sum_{p,g,j} d_{p,g,j}$$

La linéarisation est donnée par :
$$d_{p,g,j} \ge \text{occ}_{p,g,j} - T_{g,j}$$
$$d_{p,g,j} \ge -\text{occ}_{p,g,j} + T_{g,j}$$

---

#### **Problème DUAL :**

À ces contraintes sont associées deux variables duales :
$$\alpha_{p,g,j} \ge 0 \quad \text{et} \quad \beta_{p,g,j} \ge 0$$

$$-\alpha_{p,g,j} + \beta_{p,g,j} \le \lambda$$

Si on prend $\mu_{p,g,j} = \beta_{p,g,j} - \alpha_{p,g,j}$, on obtient :
$$-\lambda \le \mu_{p,g,j} \le \lambda$$

> **Observation :** On voit que $\lambda$ borne l'influence des écarts à la configuration cible.

```python
status = prob.solve()
print(pulp.LpStatus[status])

if pulp.LpStatus[status] == "Optimal":
    print(round(pulp.value(prob.objective), 2))
    for p in all_phases_inc_0:
        penalty_val = sum(pulp.value(diff_vars[p][g][j]) for g in groups for j in offices)
        print(p, round(penalty_val))
        if p > 0:
            for g in groups:
                for i in offices:
                    for j in offices:
                        val = pulp.value(move_vars[p][g][i][j])
                        if val is not None and val > 0.5:
                            print(p, g, i, j, int(round(val)))
```

# PARTIE 3 : SDP reformulation and randomized rounding

## Optimisation Dimensionnelle et Viabilité Algorithmique du Modèle SDP

### Le Défi Computationnel de la Relaxation Initiale
Une modélisation fine à l'échelle de l'individu (croisant chaque phase, chaque bureau et chaque personne) conduit à un vecteur de taille $N = 5 \times 12 \times 18 = 1080$. La matrice de décision SDP $U$ atteint alors une dimension critique de $1080 \times 1080$.

Face à cette explosion combinatoire, l'empreinte mémoire et la complexité des calculs s'avèrent rédhibitoires pour les solveurs de programmation semi-définie (aucune convergence obtenue après plus de 2 heures d'exécution).

---

### Stratégie de Réduction : L'Agrégation en Groupes Fonctionnels
Pour restaurer la viabilité du modèle tout en préservant la pertinence de l'approximation, la structure des services et les cohabitations initiales ont été exploitées. Les 18 individus sont agrégés en **12 groupes stables** (notés $e$). 

Chaque groupe $e$ possède une taille $s_e \in \{1, 2\}$ et effectue ses transitions de manière unitaire. Les regroupements stratégiques sont définis ainsi :
* **Présidence ($P$) :** Fusion des binômes des bureaux A1 et A2 en deux groupes distincts de taille 2.
* **Association Étudiante ($S$) :** Fusion des binômes des bureaux C2 et B2 en deux groupes de taille 2.
* **Techniciens ($T$) et Administratifs ($M$) :** Les paires déjà formées à l'état initial (bureau C1 pour $T$, bureau D3 pour $M$) sont traitées comme des entités indivisibles de taille 2.

Cette reformulation réduit drastiquement la taille de la matrice $U$, permettant au solveur de converger vers une solution optimale en **moins de 10 minutes**.


### Éléments du Modèle SDP

#### **Ensembles et Données**
* $E$ : ensemble des groupes de personnes, où chaque $e \in E$ représente un groupe de taille $s_e \in \{1, 2\}$.
* $Bu$ : ensemble des bureaux, $Bu = \{A1, A2, \dots, E2\}$.
* $P = \{1, \dots, 5\}$ : les phases du projet.
* $R_p \subset Bu$ : bureaux en rénovation en phase $p$.
* *Exemple :* $R_1 = \{B1, B2, B3\}$.

---

#### **Variables de Décision**
* $x_{e,b,p} \in \{0, 1\}$ :
* $1$ : si le groupe $e$ appartient au bureau $b$ en phase $p$.
* $0$ : si $e \notin b$ en phase $p$.

* $u_{e,b,p} = 2x_{e,b,p} - 1 \in \{-1, 1\}$
* $u \in \{-1, 1\}^N$ : vecteur qui regroupe tous les $u_{e,b,p}$.
* $U = u u^T \in \mathbb{R}^{N \times N}$ : matrice de corrélation SDP, avec :
* $U \succeq 0$
* $U_{ii} = 1 \quad \forall i = 1, \dots, N$

---

#### **Fonction Objectif**

L'objectif vise à minimiser les changements d'affectation entre les phases successives en exploitant les produits scalaires de la matrice $U$ :

$$\min_{U} \sum_{e,b,p} \frac{1 - U_{(e,b,p),(e,b,p+1)}}{2}$$

*Avec $p \in \{1, 4\}$

#### **Contraintes du Modèle SDP**

* **Unicité d'Affectation :**
Chaque groupe doit être affecté à un unique bureau disponible (hors rénovation) à chaque phase :
$$\sum_{b \in Bu \setminus R_p} u_{e,b,p} = 2 - |Bu \setminus R_p|$$

* **Capacité des Bureaux :**
La capacité maximale de chaque bureau (2 personnes) doit être respectée :
$$\sum_{e \in E} s_e u_{e,b,p} \le 4 - \sum_{e \in E} s_e$$

* **Contrainte de Rénovation :**
Aucun groupe ne peut occuper un bureau pendant sa phase de rénovation :
$$x_{e,b,p} = 0 \quad \forall b \in R_p, \; \forall e \in E, \; \forall p \in P$$

* **Placement Final :**
À la phase finale (phase 5), la configuration doit correspondre exactement à la cible attendue :
$$\sum_{e \in E_g} s_e x_{e,b,5} = q_{b,g}$$

*Note : $q_{b,g}$ désigne le nombre de personnes du groupe $g$ dans le bureau $b$ en phase terminale.*

### Conditions d'Optimalité de Karush-Kuhn-Tucker (KKT) pour le SDP

Pour caractériser l'optimalité de notre formulation semi-définie, nous définissons les conditions KKT associées :

#### **1. Faisabilité Primale**
La matrice de décision $U$ doit respecter les contraintes du problème primal :
* $U \succeq 0$
* $\langle A_k, U \rangle = b_k$
* $\langle A_\ell, U \rangle \le b_\ell$

---

#### **2. Faisabilité Duale**
Le multiplicateur dual (ou matrice de coût réduit) $Z$ associé à la contrainte de semi-définitude doit être semi-défini positif :
* $Z = C - \sum_{k} \nu_k A_k - \sum_{\ell} \mu_{\ell} A_{\ell} \succeq 0$
* $\mu_{\ell} \ge 0$

*Où $\nu_k$ et $\mu_{\ell}$ sont les multiplicateurs de Lagrange associés respectivement aux contraintes d'égalité et d'inégalité.*

---

#### **3. Écart Complémentaire (Complementary Slackness)**
Le produit de la matrice primale et de la matrice duale doit être nul à l'optimalité :
* $Z U = 0$

### Forme Primale

Le problème primal est de la forme :

$$\min_{U \in \mathcal{S}^N} \quad \langle C, U \rangle$$

$$\text{t.q.} \quad \langle A_k, U \rangle = b_k \quad \forall k \in E$$
$$\langle A_\ell, U \rangle \le b_\ell \quad \forall \ell \in I$$
$$U \succeq 0$$

---

### Forme Duale

#### **Variables Duales :**
* $\lambda_k \in \mathbb{R}$ pour les contraintes d'égalité.
* $\mu_\ell \ge 0$ pour les contraintes d'inégalité.
* Matrice duale $Z \in \mathcal{S}^N$, associée à la contrainte $U \succeq 0$.

#### **Problème Dual :**

$$\max_{\lambda, \mu} \quad \sum_{k \in E} \lambda_k b_k + \sum_{\ell \in I} \mu_\ell b_\ell$$

$$\text{t.q.} \quad C - \sum_{k \in E} \lambda_k A_k - \sum_{\ell \in I} \mu_\ell A_\ell = Z$$
$$Z \succeq 0$$
$$\mu_\ell \ge 0 \quad \forall \ell \in I$$

### Implémentation du Modèle SDP en Python

```python
import cvxpy as cp
import numpy as np

#on forme les groupes e, leur nombre et le placement initiale 
en = [
    {'id': 0, 'size': 2, 'type': 'P', 'start': 'A1'},
    {'id': 1, 'size': 2, 'type': 'P', 'start': 'A2'},
    {'id': 2, 'size': 1, 'type': 'O', 'start': 'B1'}, 
    {'id': 3, 'size': 1, 'type': 'M', 'start': 'B1'}, 
    {'id': 4, 'size': 2, 'type': 'S', 'start': 'B2'},
    {'id': 5, 'size': 1, 'type': 'O', 'start': 'B3'}, 
    {'id': 6, 'size': 1, 'type': 'T', 'start': 'B3'}, 
    {'id': 7, 'size': 2, 'type': 'T', 'start': 'C1'},
    {'id': 8, 'size': 2, 'type': 'S', 'start': 'C2'},
    {'id': 9, 'size': 1, 'type': 'T', 'start': 'D1'},
    {'id': 10, 'size': 1, 'type': 'M', 'start': 'D2'},
    {'id': 11, 'size': 2, 'type': 'M', 'start': 'D3'}
]

offices = ['A1','A2','B1','B2','B3','C1','C2','D1','D2','D3','E1','E2']
renovation = {1:['B1','B2','B3'],2:['D1','D2','D3'],3:['C1','C2'],4:['A1','A2'],5:[]}
#phase de 1 à 5 : la phase zero est utile uniqueme pour fixer les places des groupe e a la phase initiale 
phases = [1,2,3,4,5]

#ici on creer le triplet e,bureau,p hase
var_map = {}
idx = 1

for p in phases:
    for ent in en:
        for off in offices:
            #on ne considere pas le cas ou le bureau est en renovation
            if off not in renovation[p]:
                var_map[(ent['id'], off, p)] = idx
                idx += 1

N = idx
U = cp.Variable((N, N), PSD=True) #contraintes matrice U 
constraints = [cp.diag(U) == 1, U[0, 0] == 1] #contraintes matrice U 
obj_fct = 0

for p in phases:
    open_offices = [o for o in offices if o not in renovation[p]]
    
    #boucles pour les contraintes 
    for ent in en:
        e_idx = [var_map[(ent['id'], o, p)] for o in open_offices if (ent['id'], o, p) in var_map]
        constraints += [cp.sum([U[0,i] for i in e_idx]) == 2 - len(e_idx)]

    for off in open_offices:
        w = 0
        s = 0
        for ent in en:
            if (ent['id'], off, p) in var_map:
                u = var_map[(ent['id'], off, p)]
                w += ent['size'] * U[0,u]
                s += ent['size']
        constraints += [w <= 4 - s]

final_etats = {
    'A1': ('P', 2), 'A2': ('P', 2),
    'B1': ('M', 1), 'B2': ('M', 2), 'B3': ('M', 1),
    'C1': ('S', 2), 'C2': ('S', 2),
    'D1': ('T', 1), 'D2': ('T', 2), 'D3': ('T', 1),
    'E1': ('O', 1), 'E2': ('O', 1)
}

#boucles pour placement final 
for off,(g,q) in final_etats.items():
    idxs = [var_map[(e['id'],off,5)] for e in en if e['type']==g and (e['id'],off,5) in var_map]
    sizes = [e['size'] for e in en if e['type']==g and (e['id'],off,5) in var_map]
    constraints += [cp.sum([sizes[i]*U[0,idxs[i]] for i in range(len(idxs))]) == 2*q - sum(sizes)]

#pour limiter le nombre de déplacement// fonction objectif...
for ent in en:
    start = ent['start']
    if start not in renovation[1]:
        if (ent['id'], start, 1) in var_map:
            u = var_map[(ent['id'], start, 1)]
            obj_fct += 0.5 * (1 - U[0,u])
    else:
        obj_fct += 1

for ent in en:
    for p in range(1,5):
        for off in offices:
            if (ent['id'],off,p) in var_map and (ent['id'],off,p+1) in var_map:
                i1 = var_map[(ent['id'],off,p)]
                i2 = var_map[(ent['id'],off,p+1)]
                obj_fct += 0.5 * (1 - U[i1,i2])

prob = cp.Problem(cp.Minimize(obj_fct), constraints)
prob.solve(solver=cp.SCS, eps=1e-3)
```

### Randomized Rounding — Goemans-Williamson

Starting from the SDP solution obtained in the previous section, we apply the Goemans-Williamson randomization approach to recover an integer solution.

Since $U$ is positive semidefinite, it can be factorized as $U = VV^\top$. A random Gaussian vector is then sampled and the variables are projected onto $\{-1, 1\}$ via a sign function, yielding a binary allocation.

**Procedure :**
1. Spectral decomposition of $U$ : set negative eigenvalues to 0, compute $V = \text{eigvecs} \cdot \text{diag}(\sqrt{\text{eigvals}})$
2. Sample $r \sim \mathcal{N}(0, I)$
3. Compute projections $Vr$ and round : $u = \text{sign}(Vr)$, then $x = \frac{1 + u}{2}$

```python
# 1. Décomposition spectrale 
eigvals, eigvecs = np.linalg.eigh(U.value)
eigvals[eigvals < 0] = 0
V = eigvecs @ np.diag(np.sqrt(eigvals))

# 2. Tirage aléatoire gaussien
r = np.random.randn(V.shape[1])

# 3. Solution binaire 
projections = V @ r
u_rand = np.sign(projections) * np.sign(projections[0]) 

x_rand = (1 + u_rand) / 2


print(f"Nombre total d'affectations actives : {int(np.sum(x_rand))}")
print("\nExemples d'affectations actives (entité, bureau, phase) :")
count = 0
for key, idx in var_map.items():
    if x_rand[idx] > 0.5:
        print(key)
        count += 1
    if count >= 20: break
```

### Résultats

La solution entière obtenue après randomisation ne satisfait pas toutes les contraintes du problème — en particulier, l'unicité d'affectation par entité et par phase est violée (certaines entités se retrouvent sans bureau ou affectées à plusieurs bureaux simultanément, par exemple l'entité 11 assignée à la fois à B1 et B2 en phase 3).

Ce résultat est attendu : la procédure de Goemans-Williamson ne garantit pas la faisabilité de la solution arrondie. La solution obtenue n'est donc ni admissible ni optimale.

Pour récupérer une solution admissible à partir de ce résultat, on pourrait appliquer une étape de réparation — par exemple, réaffecter les entités avec plusieurs bureaux à leur meilleur candidat unique, ou relancer le rounding plusieurs fois et conserver la meilleure solution admissible obtenue.
